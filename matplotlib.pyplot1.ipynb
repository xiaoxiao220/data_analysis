### pyplot 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  inline
data = np.arange(0,1.1,0.01)
plt.title("lines")
plt.xlabel("the value of x")
plt.ylabel("the value of y")
plt.xlim((0,1))
plt.ylim((0,1))
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.1,0.3,0.5,0.7,0.9])
plt.plot(data,data**2)
plt.plot(data,data**4)
plt.legend(["y=x^2",'y=x^4'])
plt.savefig("./pictures/y=x^2.png")
plt.show()
```

### subplot 

```python
rad = np.arange(0,np.pi*2,0.01) # the auto value
# the first son draw
p1 = plt.figure(figsize=(8,6),dpi=80) # create draw figsize
ax1 = p1.add_subplot(2,1,1) # create 2 lines 1 raw son draw ,start first draw
# the draw property
plt.title("lines")
plt.xlabel("the value of x")
plt.ylabel("the value of y")
plt.xlim(  (0,1) ) # tuple
plt.ylim(  (0,1) )
plt.xticks(  [0,0.1,0.3,0.5,0.7,0.9]   ) # list
plt.yticks(  [0,0.2,0.4,0.6,0.8,1]     )
plt.plot(rad, rad**2) # draw function picture
plt.plot(rad, rad**4)
plt.legend(["y=x^2",'y=x^4']) # add legend 
# the second son draw 

ax2 = p1.add_subplot(2,1,2)# start second draw
plt.title("sin/cos")
plt.xlabel("rad")
plt.ylabel("value")
plt.xlim(  (0,np.pi*2) )
plt.ylim(  (-1,1) )
plt.xticks( [0,np.pi/2,np.pi,np.pi*1.5,np.pi*2])
plt.yticks( [-1,-0.5,1,0,0.5,1])
plt.plot(rad,np.sin(rad))
plt.plot(rad,np.cos(rad))
plt.legend( ['y=sin(x)','y=cos(x)'])
plt.savefig("./pictures/sincos.png")
plt.show()
```

### set rc Params 

```python
# the origin graph
x  = np.linspace(0,4*np.pi) # the value of x
y  = np.sin(x) # the value of y
plt.figure(figsize=(8,7))
plt.title("sin function")
plt.plot(x,y,label="$sin(x)$") # create draw 
plt.savefig('./pictures/default.png')
plt.show()
```

###  modify rcParams

```python
# modify rcParams graph
plt.rcParams['lines.linestyle']  = '-.'
plt.rcParams['lines.linewidth']  = 3
plt.title("modify sin function")
plt.plot(x,y,label="$sin(x)$")
plt.savefig("./pictures/modifysin.png")
plt.show()
```
